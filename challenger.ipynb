```python
#topairports
tabla=tabla.set_index('arr_port')
topairports=topairports.set_index('arr_port')
```

```python


result = pd.concat([tabla, topairports], axis=1, join='inner')
result
```

```python
result['pax_norm']= result['pax']/result['pax'].max()*100
```

```python
sample=pd.read_csv('searches.csv.bz2',sep="^", header=0,nrows=100001,na_values="NaN")

sample.describe()
```

```python
sample=sample[sample["Destination"].isin(["BCN","MAD","AGP"])]
```

```python
sample["Date"]=pd.to_datetime(sample["Date"])
sample
```

```python
def fun(x):
    return x.loc['Date'].month

def fun2(x):
    return x.loc['Date'].year

def fun3(x):
    return str(x.loc['Date'].year)+'-'+str(x.loc['Date'].month)
#sample["Date"][0].hour
#by antonio martinez
sample["Month"]=sample.apply(fun,axis=1)
sample["Year"]=sample.apply(fun2,axis=1)
sample["Year-Month"]=sample.apply(fun3,axis=1)
#sample["Year-Month"] = sample["Year-Month"].replace(r'\\n',' ', regex=True) 
```
