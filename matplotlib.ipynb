```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import pyplot as plt 
#import matplotlib as plt #alternativas bokeh plotly
```

```python
x= np.arange(0,10,.1)
y= x+np.random.normal(0,1,x.size)
y_2=np.array([x_ + np.random.normal(0,x_/2) for x_ in x])
```

```python
#pedimos figura i ejes
#representacion con linea

fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x3= np.random.beta(1,4,size=10000)
y3= np.random.beta(5,3,size=10000)
```

```python
pois = np.random.normal(3,10,1000)
pois2 = np.random.normal(20,4,500)
```

```python
x4=np.arange(-2,2,0.05)
y4= x4**3
```

```python
import seaborn as sns #mapas de calor
```

```python
#ponemos puntos
fig,ax=plt.subplots(6,2,figsize=(13,25))
ax[0,0].plot(x,y,'--D') #lineas '--' disconituna matplot lib markers
ax[0,0].plot(x,y_2,linewidth=4) #lineas
ax[0,1].scatter(x,y) #puntitos
ax[0,1].scatter(x,y_2,s=10,c='r') #puntitos s=tamao puntos c='r' color rojo #le podemos pasar un array 
ax[1,0].scatter(x3,y3,alpha=0.1,s=1) #alpha hace transparente
ax[1,1].hist(pois,bins=30) #histograma (bins=numero de barras)
ax[1,1].hist(pois2,bins=30,alpha=0.5) #histograma (bins=numero de barras)
ax[2,0].plot(x4,y4)
ax[2,0].plot(np.arange(0.01,2,0.05),np.log(np.arange(0.01,2,0.05))) #el valor y una funcion
ax[2,1].plot(x4,np.sin(x4))
ax[2,1].plot(x4,np.cos(x4))
ax[2,1].grid() ##aade una rejilla
ax[3,0].hist(pois,cumulative=True,normed=True,bins=100) #hisrograma tuneado se llama CDF va acumulando
ax[3,1].bar([0,1,2,3],[3,5,1,10])
ax[4,0].boxplot(pois)
sns.heatmap([[1,2,2,10],[2,3,1,4],[2,1,1,1],[8,6,3,1]],ax=ax[4,1],annot=True) #mapa de calor
sns.violinplot(pois,ax=ax[5,0])
sns.jointplot(x,y_2)
plt.show()

#ohlc
```
